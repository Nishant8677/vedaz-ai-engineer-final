# Phase 2: Dataset Analysis

In this notebook, we analyze the conversational dataset for the Qwen fine-tuning project.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
data_path = '../data/processed/cleaned_dataset.jsonl'
conversations = []
with open(data_path, 'r', encoding='utf-8') as f:
    for line in f:
        conversations.append(json.loads(line))
```

```python
num_chats = len(conversations)
print(f"Total conversations: {num_chats}")
```

```python
lengths = []
roles = []
sys_prompts = set()
missing_fields = 0

for chat in conversations:
    messages = chat.get('messages', [])
    lengths.append(len(messages))
    for msg in messages:
        if 'role' not in msg or 'content' not in msg:
            missing_fields += 1
        roles.append(msg.get('role', 'unknown'))
        if msg.get('role') == 'system':
            sys_prompts.add(msg.get('content'))

avg_length = sum(lengths) / len(lengths) if lengths else 0
print(f"Average messages per conversation: {avg_length:.2f}")
print(f"Unique system prompts: {len(sys_prompts)}")
print(f"Total missing fields (role/content): {missing_fields}")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(lengths, bins=10, kde=True)
plt.title('Distribution of Conversation Lengths')
plt.xlabel('Number of Messages')
plt.ylabel('Frequency')
plt.show()
```

```python
role_counts = pd.Series(roles).value_counts()
plt.figure(figsize=(6,4))
sns.barplot(x=role_counts.index, y=role_counts.values)
plt.title('Role Counts in Dataset')
plt.xlabel('Role')
plt.ylabel('Count')
plt.show()
```
